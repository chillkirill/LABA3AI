```python
import numpy as np
import matplotlib.pyplot as plt

matrix = np.random.rand(10, 10)

#  матрица как тепловая карта
plt.imshow(matrix, cmap='viridis', interpolation='nearest')  #  цветовая палитра
plt.colorbar(label='Значение')  #  цветовая шкала

plt.title('Тепловая карта случайной матрицы 10x10')
plt.xlabel('Столбец')
plt.ylabel('Строка')
plt.xticks(np.arange(10))  #  все отметки на оси x
plt.yticks(np.arange(10))  #  все отметки на оси y

plt.show()
```
