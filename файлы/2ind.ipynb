```python
import matplotlib.pyplot as plt

days = ['Пн', 'Вт', 'Ср', 'Чт', 'Пт', 'Сб', 'Вс']
visitors = [150, 180, 200, 220, 300, 450, 500]

#  цвета для будних и выходных дней
colors = ['blue'] * 5 + ['green'] * 2  # Первые 5 дней - будние (синий) последние 2 - выходные (зеленый)

#  столбчатая диаграмма
plt.figure(figsize=(8, 6))  
plt.bar(days, visitors, color=colors)

plt.xlabel('День недели')
plt.ylabel('Количество посетителей')
plt.title('Количество посетителей за неделю')

plt.show()
```
