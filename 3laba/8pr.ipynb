```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 100)  #  100     -5  5

fig, axes = plt.subplots(2, 2, figsize=(10, 8)) #  2x2

#    y = x
axes[0, 0].plot(x, x)
axes[0, 0].set_title('  y = x')
axes[0, 0].set_xlabel('x')
axes[0, 0].set_ylabel('y')
axes[0, 0].grid(True)

#   y = x^2
axes[0, 1].plot(x, x**2)
axes[0, 1].set_title(' y = x^2')
axes[0, 1].set_xlabel('x')
axes[0, 1].set_ylabel('y')
axes[0, 1].grid(True)

#   y = sin(x)
axes[1, 0].plot(x, np.sin(x))
axes[1, 0].set_title(' y = sin(x)')
axes[1, 0].set_xlabel('x')
axes[1, 0].set_ylabel('y')
axes[1, 0].grid(True)

#   y = cos(x)
axes[1, 1].plot(x, np.cos(x))
axes[1, 1].set_title(' y = cos(x)')
axes[1, 1].set_xlabel('x')
axes[1, 1].set_ylabel('y')
axes[1, 1].grid(True)

#   
plt.tight_layout()

plt.show()
```
