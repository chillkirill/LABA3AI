```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 400)  #  интервал [-5, 5] 
y1 = x
y2 = x**2
y3 = x**3

plt.figure(figsize=(8, 6))

# Построение линий 
plt.plot(x, y1, 'b:', label='y = x')  # Синяя пунктирная линия
plt.plot(x, y2, 'g-.', label='y = x²') # Зеленая штрих-пунктирная линия
plt.plot(x, y3, 'r-', label='y = x³')  # Красная сплошная линия

plt.xlabel('x')
plt.ylabel('y')
plt.title('Графики функций y = x, y = x², y = x³')
plt.grid(True)
plt.legend()  

# Одинаковый масштаб осей:
plt.axis('equal') 

plt.show()
```
