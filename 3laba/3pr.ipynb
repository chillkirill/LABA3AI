```python
import matplotlib.pyplot as plt
import numpy as np

n = 50  
x = np.random.rand(n) * 20 - 10  # x в диапазоне [-10, 10]
y = np.random.rand(n) * 20 - 10  # y в диапазоне [-10, 10]

colors = x  
sizes = y * 10 # умножение для наглядности

plt.figure(figsize=(8, 6))  #  размер фигуры 
plt.scatter(x, y, c=colors, s=sizes, alpha=0.7, cmap='viridis')  

plt.xlabel('x')
plt.ylabel('y')
plt.title('Диаграмма рассеяния: Цвет от x, Размер от y')
plt.colorbar(label='Значение x')  #  цветовая шкала для цвета
plt.grid(True)

plt.show()

```
